To start exploring the 7th SDG, let us load the file using the [`read_csv`](https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html) function.

As we can see, for this SDG, we only have one target included in the dataset. 

First, let us determine the `Year` values of our data, so that we can decide how to visualize it.

```python
print ('Year values: ', sdg7 ['Year'].unique())
print ('Number of years: ', len(sdg7 ['Year'].unique()))
```

From this, we can see that we have 12 years of `7.1.1 Proportion of pop with electricity` data that we have to explore.

From the **Data Preparation Notebook**, we remember dropping the **NaN** rows, but to be sure, let us check how many **NaN** values it has using the [`isnull`](https://pandas.pydata.org/docs/reference/api/pandas.isnull.html) and [`sum`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sum.html) functions.

#### Per region, which year has the lowest `Proportion of the population with access to electricity`?
To answer this question, let us first get the regions that are included in the dataset. This would help us determining which regions would have an answer for this question.

```python
geolocation_vals = sdg7 ['Geolocation'].unique ()
geolocation_vals
```

```python
len(geolocation_vals)
```

So, we would have to find the which year has the lowest value for the one target in this SDG for 18 different `Geolocation` values.

However, from the returned unique values of `Geolocation`, we know that the **PHILIPPINES** is not a region of the country, but it is the overall `7.1.1 Proportion of pop with electricity`. To avoid confusion, let us remove it from the list of regions that we would explore later.

```python
geolocation_vals = geolocation_vals [1: ]
```

Now, we will explore the rows of the **PHILIPPINES** first. As we want to check the year that has the lowest value for the **PHILIPPINES**, we will be sorting it in ascending order using the [`sort_values`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sort_values.html) function.

```python
philippines = sdg7 [sdg7 ['Geolocation'] == 'PHILIPPINES']
philippines = philippines.sort_values(by='7.1.1 Proportion of pop with electricity', ascending = True)
philippines
```

Based on the output, we can see that in the year **2011**, Philippines logged its lowest proportion of the population with access to energy. 

However, as it is hard to visualize it with just the numbers, let us use a line graph to see the trend.

```python
philippines = philippines.sort_values(by='Year', ascending = True)
plt.plot(philippines ['Year'], philippines ['7.1.1 Proportion of pop with electricity'])
plt.title('PHILIPPINES')
plt.ylim(0, 120)
plt.show()
```

##### Figure 4.1. Line Graph of the Proportion of the Population with access to Electricity in the Philippines
From Figure 4.1., we can see that from a high value of proportion in 2009 and 2010, the value dipped to its lowest in 2011. It has been increasing from 2011 to 2020. Although there are years that it decreases, it was not as big as the decrease from 2010 to 2011.

As we know that the proportion of the **PHILIPPINES**, as a whole, is affected by the proportions in the region. Let us now explore the regions to check if the values also decreased in 2011. 

```python
# Declaring 17 subplots for the 17 `Geolocation` values that we would need to plot
figure, axis = plt.subplots((len (geolocation_vals) + 1) // 2, 2, figsize = (15, 30))
max_ = sdg7['7.1.1 Proportion of pop with electricity'].max()
min_ = sdg7['7.1.1 Proportion of pop with electricity'].min()
# Setting the range of values for the y-axis of each subplot
for axis_ in axis:
    for axis__ in axis_:
        axis__.set_ylim([min_ - 10, max_ + 10])

# Creating the graph for each of the `Geolocation` values
counter = 0
for i in range ((len (geolocation_vals) + 1) // 2):
    for j in range (2):
        if counter < len(geolocation_vals):
            temp = sdg7 [sdg7 ['Geolocation'] == geolocation_vals [counter]]
            axis[i, j].plot(temp ['Year'], temp['7.1.1 Proportion of pop with electricity'])
            axis[i, j].set_title(geolocation_vals [counter])
            counter = counter + 1
        else:
            plt.delaxes(axis[i, j])
            
# Showing the outpuit
plt.subplots_adjust(left = 0.1, bottom = 0.1, right = 0.9, top = 0.9, wspace = 0.4, hspace = 0.4)
plt.show()
```

##### Figure 4.2. Line Graph of the Proportion of the Population with access to Electricity for each of the Regions in the Philippines
From these subplots, we can see that for all of the regions, there is a sudden decrease for the year **2011**. This is why we can easily see that this is year holds the lowest value for most of the regions. However, for **Region 10: Northern Mindanao**, we can see the it became lower in 2013.

### Which indicators have the largest correlations with each other?
To determine the correlations of each of the indicators with each other, we can use the [`corr`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.corr.html) function on the columns that are for the datasets. 

```python
correlation = data.loc[ : , '1.2.1 Poverty Proportion' : ].corr()
correlation
```

From the data frame above, we can see that there are **1.0** values in the diagonal line, which is the correlation of a column with itself. Simply getting the pairs that has the highest correlation would return these **1.0** values, which is why we have to remove either the lower or upper half of the data frame.

```python
corr_upper_df = correlation.where(np.triu(np.ones(correlation.shape), k=1).astype(bool))
corr_upper_df
```

After removing the lower half, we can now unstack it, drop the pairs that are the same, and turn it into a dataframe. We would not be immediately sorting the data frame as we want to find those columns that are positive correlated and those that are negatively correlated.

```python
corr_pairs = pd.DataFrame(corr_upper_df.unstack().sort_values().drop_duplicates())
corr_pairs.columns = ['Correlation values']
corr_pairs
```

Once we have created a data frame that holds the pairs of columns and their correlation, we can now sort it in ascending order based on the `Correlation values`. This would help us see the columns that has the highest positive correlation.

```python
corr_pairs = corr_pairs.sort_values (by = 'Correlation values', ascending = False)
corr_pairs [0: 5]
```

The top five pairs of columns that has the highest positive correlations are: (1) `14.5.1.2 Coverage of protected NIPAS and Locally managed MPAs` and `14.5.1.1 Coverage of protected areas`, (2) `GRDP` and `Gross Capital Formation`, (3) `Gross Capital Formation` and `4.c TVET trainers trained`, (4) `Population` and `Consumption Expenditure %`, and (5) `Population` and `4.c TVET trainers trained`. These means that they both as the value in one of the columns increases/decreases, the other also increases/decreases (i.e., the grow in the same way).

Next, we can look at those columns that are negatively correlated.

```python
corr_pairs = corr_pairs.sort_values (by = 'Correlation values', ascending = True)
corr_pairs [0: 5]
```

These are the top five columns that are negatively correlated with each other: (1) `Drop-out rate` and `4.1 Elem Completion Rate`, (2) `1.4.1 Net SHS Enrolment Rate` and `1.2.1 Poverty Proportion`, (3) `1.4.1 Net JHS Enrolment Rat`e and `1.2.1 Poverty Proportion`, (4) `Drop-out rate` and `3.4.1 Mortality rate creditted to NCD`, and (5) `3.4.1 Mortality rate creditted to NCD` and `1.2.1 Poverty Proportion`. As these columns are negatively correlated with each other, this means that as one of this columns increases/decreases, the other column in the pair decreases/increases.

We can visualize their correlation with each other using a heatmap.

```python
plt.figure (figsize = (50, 50))
sns.set(font_scale = 2.4)
heatpmap = sns.heatmap (data = correlation, xticklabels = correlation.columns.values, yticklabels = correlation.columns.values, cmap = sns.diverging_palette(20, 220, n=200), annot=True)
plt.show()
```

##### Figure 9.2. Half Heatmap of the Different Columns
Using the half heatmap, it's now easier to see which colums have high correlations. 

As the goal of this dataset, let us focus on the interesting correlations between the different SDGs. We can see that there is a strong negative correlative between `Poverty Population` and the following: (1) `Net JHS Enrolment Rate`, (2) `Net SHS Enrolment Rate`, (3) `Mortality Rate attributeed to NCD`, and (4) `Proportion of Population with access to Electicity`. This means that as Poverty Population increases (or decreases), these other indicators decreases (or increases), or vice versa.

We can also see that `Coverage of protected areas` is highly positive correlated with the `Coverage of protected NIPAS and Locally managed MPAs`, as they have a correlation of **1**. This might be because they are both under the same target of an SDG.

Another interesting correlation is the high positive correlation between the `Population` and `Coverage of protected NIPAS and Locally managed MPAs`. This means as the population increases, the million hectares of total sea area that is protected also increases, or vice versa.

There is also a high negative correlation between the `Drop-out rate` and the `Mortality rate credited to NCD`. As it's a negative correlation, this means that as the drop-out rate decreases (or increases), the mortality rate credited to cardiovascular disease, cancer, diabetes or chronic respiratory disease increases (or decreases), or vice versa. 

Additionally, we can see that there is a high positive correlation between the number of `TVET trainers trained` and the `Gross Capital Formation`. Thus, this means that as one of these two values increases, the other also increases.
